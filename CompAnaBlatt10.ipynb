# Aufgabe 1

```python
r,p,t=symbols('r p t')
Phi=Matrix(3,1,[r*cos(p)*cos(t),r*sin(p)*cos(t),r*sin(t)])
lsg=solve(det(Phi.jacobian([r,p,t])))
lsg
```

```python
Phi.subs(lsg[0])
```

```python
Phi.subs(lsg[1])
```

```python
Phi.subs(lsg[2])
```

```python
# Zusammengefasst also alle Punkte auf der z-Achse.
```

# Aufgabe 2

```python
x,y=symbols('x y')
f=x*(x**2-1)*exp(-x**2-y**2-y)
gr=Matrix([f]).jacobian([x,y])
H=hessian(f,[x,y])
krit=solve(gr)
krit
```

```python
H.subs(krit[0]).is_negative_definite
```

```python
H.subs(krit[1]).is_positive_definite
```

```python
H.subs(krit[2]).is_positive_definite
```

```python
H.subs(krit[3]).is_negative_definite
```

```python
# Maximalstellen bei Lösung 1 und 4 und Minimalstellen bei Lösung 2 und 3.
```

# Aufgabe 3

```python
a,r,h=symbols('a r h',positive=true)
f=2*pi*r*h+2*a*pi*r**2
g=pi*r**2*h-1
solve(g)
```

```python
f.subs(solve(g)[0])
```

```python
solve(f.subs(solve(g)[0]).diff(r),r)
```

```python
solve(f.subs(solve(g)[0]).diff(r),r)[0],1/(pi*solve(f.subs(solve(g)[0]).diff(r),r)[0]**2)
```

```python
# Minimalstelle nach Satz von Rolle - f wird für kleine und große r unter g=0 beliebig groß, ist aber stetig.
```

```python
# Die gesuchte Potenz ist 2/3.
```

# Aufgabe 4

```python
tn = np.linspace(-np.pi, np.pi, 300)
xn = np.cos(tn)
yn = np.sin(tn)
plt.plot(xn, yn)
plt.axis('equal');
```

```python
# Das funktioniert, indem Sinus gegen Kosinus geplottet wird (-> komplexe Polardarstellung). 
# Die letzte Zeile sorgt dafür, dass die rechteckige Darstellung nichts verzerrt.
```

```python
tn = np.linspace(0, 100, 1000)
xn = np.cos(tn)*tn**(1/2)
yn = np.sin(tn)*tn**(1/2)
plt.plot(xn, yn)
plt.plot(-xn, -yn)
plt.axis('equal');
```
